**13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

  The CLt states that if you have a population with a mean and standard deviation and take sufficiently large no of random samples from the population
  with replacement then the distribution of sample means will be approximately normal distribution. Regardless of the original distribution of a population (normal or not), the sampling distribution of the sample mean will approach a normal distribution as the sample size (n) becomes large. It is important as
  enables Use of Normal-Based Methods:
  It allows us to apply normal distribution techniques (like Z-tests, confidence intervals) to non-normal data, as long as the sample size is large enough.
  Forms the Basis of Inference:
  It justifies why we can make conclusions about a population based on sample statistics (like the mean).
  Works for Many Types of Distributions:
  Whether the population is exponential, uniform, or binomial — the sample means will still approximate a normal distribution.
  Practical in Real Life:
  Most real-world problems use samples, not entire populations. CLT makes it possible to make accurate predictions and decisions based on those samples.

**14.How does the Central Limit Theorem relate to the normal distribution?**

  The Central Limit Theorem (CLT) is directly related to the normal distribution because it explains why and when the normal distribution appears in statistics, even when the data isn't normal to begin with.
  They Relate:
  CLT → Normal Distribution Emerges from Sampling:
  The CLT states that the sampling distribution of the sample mean becomes approximately normal as the sample size increases, no matter what the population distribution looks like (skewed, uniform, etc.).
  Shape of Sampling Distribution Becomes Normal:
  Even if individual data points come from a non-normal population, the distribution of their averages will resemble a normal distribution.
  Foundation for Normal-Based Statistical Methods:
  Because of the CLT, we can:
  Use normal distribution-based formulas (like z-scores or confidence intervals) for means.
  Justify using normal probability models in many real-world situations, especially with large datasets.
  
**15.What is the application of Z statistics in hypothesis testing?**

  Z-statistics (or Z-scores) are widely used in hypothesis testing when the population parameters (especially the standard deviation) are known or the sample size is large n≥30.

**16.How do you calculate a Z-score, and what does it represent?**

  Z-score = X-μ/σ
  Z=0: The value is exactly at the mean.
  Z=+1: The value is 1 standard deviation above the mean.
  Z=-2: The value is 2 standard deviations below the mean.
  The higher the absolute value of Z, the further the data is from the mean.

**17.What are point estimates and interval estimates in statistics?**

  A point estimate is a single value that serves as a best guess of an unknown population parameter.
  An interval estimate gives a range of values within which the population parameter is likely to lie, along with a confidence level.

**18.What is the significance of confidence intervals in statistical analysis?**

  A confidence interval (CI) is crucial in statistics because it provides a range of possible values for an unknown population parameter — rather than just a single estimate — and includes a confidence level that quantifies the uncertainty.

**19.What is the relationship between a Z-score and a confidence interval?**

  The Z-score plays a central role in calculating confidence intervals when the population standard deviation is known or the sample size is large. A confidence interval is calculated using the Z-score that corresponds to the desired confidence level.
  The Z-score determines the width of the confidence interval.
  Higher confidence → larger Z → wider interval (more certainty but less precision).
  Confidence intervals are built from Z-scores to quantify the uncertainty around a sample estimate

**20.How are Z-scores used to compare different distributions?**

  Z-scores allow us to standardize values from different distributions so they can be compared on the same scale, even if the original data have different means and standard deviations. Z-scores convert raw values into a common scale: the number of standard deviations from the mean.

**21.What are the assumptions for applying the Central Limit Theorem?**

  The assumptions of the CLT:
  Random Sampling-: Samples must be randomly selected from the population to avoid bias.
  Independence of Observations-: Each observation should be independent of the others.If sampling without replacement, the sample size should be less than 10% of the population to maintain independence.
  Sample Size-: A large enough sample size is needed, especially if the population is not normally distributed. sample size n≥30 for most distributions.
  Finite Variance (σ²)-: The population must have a finite standard deviation. Infinite or undefined variance breaks the CLT.

**22.What is the concept of expected value in a probability distribution?**

  The expected value (also called mean or mathematical expectation) of a random variable is the long-run average outcome you'd expect if an experiment were repeated many times. In simple terms, the expected value is a weighted average of all possible values, where each value is weighted by its probability.

**23.How does a probability distribution relate to the expected outcome of a random variable?**

  A probability distribution defines how likely each possible outcome of a random variable is, and the expected outcome (or expected value) is the average result you'd expect over many repetitions — based on that distributionThe probability distribution assigns probabilities to all possible values of the random variable.
  The expected value is calculated by weighing each value by its probability — essentially, a weighted average.
